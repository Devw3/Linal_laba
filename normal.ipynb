```python
from sage.all import *
import random
import math
```

```python
isu = 501743
N = isu % 20

m = 4 + (N % 5)
n = 2 + (N % 10)
k = 1 + (N % 7)

n1 = N % 6
n2 = (N + 1) % 6
n3 = (N + 2) % 6

p = 23
s = 17
r = 45
t = 12

G = SymmetricGroup(m)

def subgroups_of_Sm(N: int) -> dict:
    subs = G.subgroups()
    num_subs = len(subs)
    random_sub = random.choice(subs)
    idx = N % num_subs
    H = subs[idx]
    left_cosets = G.cosets(H, side="left")
    right_cosets = G.cosets(H, side="right")
    index = G.order() / H.order()
    is_normal = H.is_normal(G)
    return {
        'number_of_subgroups': num_subs,
        'random_subgroup': random_sub,
        'selected_subgroup': H,
        'left_cosets': left_cosets,
        'right_cosets': right_cosets,
        'index': index,
        'is_normal': is_normal
    }

result = subgroups_of_Sm(N)
print("количество подгрупп:", result['number_of_subgroups'])
print("случайная подгруппа:", result['random_subgroup'])
print("выбранная подгруппа:", N % result['number_of_subgroups'], "):", result['selected_subgroup'])
print("левые смежные классы:", result['left_cosets'][:10])
print("правые смежные классы:", result['right_cosets'][:10])
print("индекс:", result['index'])
print("нормальная или нет:", result['is_normal'])
```

```python
def element_powers_in_Sm(N: int) -> dict:
    order_G = G.order()
    g_index = N % order_G
    elements = G.list()
    g = elements[g_index]
    
    h1 = g ** n1
    h2 = g ** n2
    h3 = g ** n3
    
    order_h1 = h1.order()
    order_h2 = h2.order()
    order_h3 = h3.order()
    
    # порядок циклической подгруппы равен порядку элемента
    subgroup_order_h1 = order_h1
    subgroup_order_h2 = order_h2
    subgroup_order_h3 = order_h3
    
    return {
        'g': g,
        'h1': h1,
        'order_h1': order_h1,
        'subgroup_order_h1': subgroup_order_h1,
        'h2': h2,
        'order_h2': order_h2,
        'subgroup_order_h2': subgroup_order_h2,
        'h3': h3,
        'order_h3': order_h3,
        'subgroup_order_h3': subgroup_order_h3
    }

result = element_powers_in_Sm(N)

print("элемент g", result['g'])
print(f"g^{n1}, n1={n1}:", result['h1'])
print("порядок g^{n1}:", result['order_h1'])
print("порядок циклической подгруппы <g^{n1}>:", result['subgroup_order_h1'])
print(f"g^{n2}, n1={n2}:", result['h2'])
print("порядок g^{n2}:", result['order_h2'])
print("порядок циклической подгруппы <g^{n2}>:", result['subgroup_order_h2'])
print(f"g^{n3}, n1={n3}:", result['h3'])
print("Порядок g^{n3}:", result['order_h3'])
print("Порядок циклической подгруппы <g^{n3}>:", result['subgroup_order_h3'])
```

```python
def solve_sigma_power_eq(N: int) -> dict:
    cycle_str = '(' + ','.join(str(i) for i in range(1, m)) + ')'
    rho = G('(' + ','.join(map(str, range(1, m))) + ')')
    solutions = [sigma for sigma in G if sigma**n == rho]
    num_solutions = len(solutions)
    random_solutions = random.sample(solutions, min(3, num_solutions))
    cycle_types = set(sigma.cycle_type() for sigma in solutions)
    common_description = f"все решения имеют цикловые тип: {cycle_types}. Они являются элементами, чья n-ая степень (n={n}) равна заданному (m-1)-циклу."
    return {
        'number_of_solutions': num_solutions,
        'random_solutions': random_solutions,
        'common_description': common_description
    }

result = solve_sigma_power_eq(N)
print("количество решений:", result['number_of_solutions'])
print("случайные решения:")
for sol in result['random_solutions']:
    print(sol)
print("общее у них:", result['common_description'])
```

```python
# определение циклической подгурппы
def generate_set_by_element(a, M):
    S = {1}
    x = a
    while x not in S:
        S.add(x)
        x = (x * a) % M
    return tuple(sorted(S))
    
# поиск всех подгрупп
def subgroups_of_Zm_star(N: int) -> list:
    m = 4 + (N % 5)
    unique_subgroups = {
        generate_set_by_element(g, m)
        for g in range(1, m)
        if math.gcd(g, m) == 1
    }
    return sorted([list(t) for t in unique_subgroups], key=lambda h: (len(h), h))

print(subgroups_of_Zm_star(501743))
```

```python
def order_and_primitivity_of_t(N: int) -> dict:
    p = 23
    t = 12
    
    # кольцо
    R = IntegerModRing(p)
    # элемент t в группе
    t_mod = R(t)
    # порядок элемента t
    order = t_mod.order()
    # проверка является ли t примитивным корнем
    phi_p = euler_phi(p)
    is_primitive = (order == phi_p)
    
    return {
        'p': p,
        't': t,
        'order': int(order),
        'phi_p': int(phi_p),
        'is_primitive': is_primitive
    }


result = order_and_primitivity_of_t(N)

# print(f"элемент t = {t}, порядок {result['order']}, phi({result['p']}) = {result['phi_p']}")
print(f"t примитивный корень? {result['is_primitive']}")
```

```python
def generators_of_Zm_star(N: int) -> list:
    m = 4 + (N % 5)
    phi = euler_phi(m)
    
    primitives = []
    for g in range(1, m):
        if gcd(g, m) == 1:
            # если порядок g равен φ(m), тогда g примитивный
            if pow(g, phi, m) == 1 and all(pow(g, phi // p, m) != 1 for p, _ in factor(phi)):
                primitives.append(g)
    
    return sorted(primitives)

print(generators_of_Zm_star(N))
```

```python
def cyclic_subgroup_in_Zm_additive(N: int) -> dict:
    # порядок подгруппы
    d = gcd(t, m)
    order = m // d

    # подгруппа по mod m
    subgroup = [(k * t) % m for k in range(order)]

    # порождающие элементы
    generators = set(
        k % m for k in range(m)
        if (k * t) % m in subgroup and gcd(k, order) == 1
    )

    return {
        'm': m,
        't': t,
        'subgroup': subgroup,
        'order': order,
        'generators': generators
    }

res = cyclic_subgroup_in_Zm_additive(N)
print(f"порядок подгруппы: {res['order']}")
print(f"подгруппа {res['t']}: {res['subgroup']}")
print(f"порождающие: {res['generators']}")
```

```python
def isomorphism_of_cyclic_subgroup_Zm_star(N: int) -> dict:
    t = 12

    if gcd(t, m) != 1:
        t = 1

    # поиск порядка элемента
    phi = euler_phi(m)
    order = 1
    pow_t = t
    while pow_t != 1:
        pow_t = (pow_t * t) % m
        order += 1
        if order > phi:
            order = 1
            break

    subgroup = [(pow(t, k, m)) for k in range(order)]
    return {
        'm': m,
        't': t,
        'subgroup': subgroup,
        'order': order
    }

res = isomorphism_of_cyclic_subgroup_Zm_star(N)
print(f"подгруппа {res['t']} = {res['subgroup']}")
print(f"Изоморфна: {[i for i in range(1, res['order']+1)]} в S_{res['order']}")
```

```python
def find_polynomial_roots_normal(N: int) -> dict:
    a = [(i + N) % 4 for i in range(9)]
    b = [(j + N) % 7 for j in range(7)]

    # поле F4
    F4 = GF(4, 'a')
    R4.<x> = PolynomialRing(F4)
    f1_poly = x^9 + sum(F4(a[i]) * x^i for i in range(9))
    roots1 = f1_poly.roots()
    roots1_list = [r[0] for r in roots1]

    # поле F7
    F7 = GF(7)
    R7.<x> = PolynomialRing(F7)
    f2_poly = sum(F7(b[j]) * x^j for j in range(7))
    roots2 = f2_poly.roots()
    roots2_list = [r[0] for r in roots2]

    return {
        'N': N,
        'field_F4': {
            'polynomial': f1_poly,
            'coefficients_a': a,
            'roots': roots1_list,
            'roots_count': len(roots1_list)
        },
        'field_F7': {
            'polynomial': f2_poly,
            'coefficients_b': b,
            'roots': roots2_list,
            'roots_count': len(roots2_list)
        }
    }

res = find_polynomial_roots_normal(N)

print(f"корни поля F_4[x]: {res['field_F4']['roots']}")
print(f"корни поля F_7[x]: {res['field_F7']['roots']}")
```

```python
def gcd_linear(N):
    r = [(i + N) % 11 for i in range(8)]
    s = [(i + N) % 11 for i in range(4)]

    F = GF(11)
    R.<x> = PolynomialRing(F, 'x')
    f = sum(F(r[i]) * x^i for i in range(8))
    g = sum(F(s[i]) * x^i for i in range(4))

    d = f.gcd(g)
    bezout = xgcd(f, g)
    _, u, v = bezout

    print("f(x) =", f)
    print("g(x) =", g)
    print("gcd(f,g) =", d)
    print(f"линейное представление: {d} = ({u}) * f(x) + ({v}) * g(x)")

N = 3
gcd_linear(N)
```

```python
def inverse_mod_g(N):
    s0 = (0 + N) % 13
    s1 = (1 + N) % 13
    s2 = (2 + N) % 13

    F = GF(13)
    R.<x> = F[]
    f = F(s2)*x^2 + F(s1)*x + F(s0)
    g = x^8 + x^4 + x^3 + F(6)*x + 2
    d, u, v = xgcd(f, g)
    
    if d != 1:
        print("f и g должны быть взаимно простыми")
        return

    h = u % g
    print("f(x) =", f)
    print("g(x) =", g)
    print("обратный элемент: h(x) =", h)
    print(f"проверка: f*h mod g = {f*h % g}")

inverse_mod_g(N)
```

```python
def generate_irreducible_polynomials(q: int, d: int) -> list:
    F = GF(q)
    R.<x> = PolynomialRing(F)

    # список всех неприводимых палиномов
    irreducibles = []
    for coeffs in cartesian_product_iterator([F for _ in range(d)]):
        poly = x^d + sum(F(coeffs[i]) * x^i for i in range(d))
        if poly.is_irreducible():
            irreducibles.append(poly)
    
    return sorted(irreducibles, key=str)

for q in [2, 3, 5]:
    for d in [2, 3, 4]:
        polys = generate_irreducible_polynomials(q, d)
        print(f"\nF_{q}, степень {d}: {len(polys)} неприводимых")
        for p in polys:
            print("  ", p)
```
